#where clause

```python
condition = (movies['Director'] == "Aamir Khan")

condition.tail(10)
```

```python
movies[movies['Director'] == "Aamir Khan"]
```

```python
movies[movies['Rating'] >= 5].head(3)
```

```python
movies[(movies['Director'] == 'Aamir Khan') | (movies['Rating'] >= 8)].tail()
```

```python
movies[movies['Director'].isin(['Aamir Khan', 'Ridley Scott'])].tail()
```

```python
movies[
    ((movies['Year'] >= 2005) & (movies['Year'] <= 2010))
    & (movies['Rating'] > 8.0)
    & (movies['Revenue'] < movies['Revenue'].quantile(0.25))
]
```

```python
def rating_function(x):
    if x >= 8.0:
        return "good"
    else:
        return "bad"
```

```python
movies["Rating_category"] = movies["Rating"].apply(rating_function)

movies.head(2)
```

```python
movies["Rating_category"] = movies["Rating"].apply(lambda x: 'good' if x >= 8.0 else 'bad')

movies.head(2)
```
